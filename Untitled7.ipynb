```python
# Import required libraries
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (update the file path accordingly)
file_path = '/content/augmented_no_improvement_Binaural_Beats_University_Students_Dataset.xlsx'
df = pd.read_excel(file_path)

# Step 1: Create Target Variables for Improvement
# For Mental Health Improvement: 1 if Mental_Habit_After > Mental_Habit_Before, else 0
df['Mental_Health_Improvement'] = (df['Mental_Habit_After'] > df['Mental_Habit_Before']).astype(int)

# For Sleep Schedule Improvement: 1 if Sleep_Habit_After > Sleep_Habit_Before, else 0
df['Sleep_Schedule_Improvement'] = (df['Sleep_Habit_After'] > df['Sleep_Habit_Before']).astype(int)

# Step 2: Select Features and Target Variable
X = df[['Age', 'Gender', 'Sleep_Habit_Before', 'Mental_Habit_Before', 'Performance_Before',
        'Creativity_Before', 'Well_Being_Before']]

# Encode categorical variables (Gender)
le = LabelEncoder()
X.loc[:, 'Gender'] = le.fit_transform(df['Gender'])

# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Ensure target is binary and properly encoded
y_mental_health = df['Mental_Health_Improvement']

# Check the unique values in the target variable
print(y_mental_health.unique())  # This will show if there are non-binary values

# Encode target variable as binary (if needed)
y_mental_health = le.fit_transform(y_mental_health)

# Step 3: Split the data into training and test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_mental_health, test_size=0.2, random_state=42, stratify=y_mental_health)

# Step 4: Train XGBoost Classifier with a proper base score (between 0 and 1)
model = xgb.XGBClassifier(eval_metric='logloss', base_score=0.5, random_state=42)

# Fit the model and track training/validation performance
# --- BEGIN MODIFICATION ---
history = model.fit(X_train, y_train,
                    eval_set=[(X_train, y_train), (X_test, y_test)], # Removed string labels from eval_set tuples
                    verbose=True)
# --- END MODIFICATION ---

# Step 5: Make predictions
y_pred = model.predict(X_test)

# Step 6: Evaluate the Model

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Classification Report
print("\nClassification Report for Mental Health Improvement:\n", classification_report(y_test, y_pred))

# Confusion Matrix for Mental Health Improvement
cm_mental_health = confusion_matrix(y_test, y_pred)
sns.heatmap(cm_mental_health, annot=True, fmt='d', cmap='Blues', xticklabels=[0, 1], yticklabels=[0, 1])
plt.title('Confusion Matrix for Mental Health Improvement')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Step 7: Plot Training and Validation Accuracy & Loss

# Plot accuracy
plt.figure(figsize=(15, 5))

# Accuracy plot (changed to plot logloss as 'accuracy' was not recorded by eval_metric='logloss')
plt.subplot(1, 2, 1)
# --- BEGIN MODIFICATION ---
plt.plot(history.evals_result()['validation_0']['logloss'], 'o-', label='Training Logloss') # Access training logloss via validation_0
plt.plot(history.evals_result()['validation_1']['logloss'], 'o-', label='Validation Logloss') # Access validation logloss via validation_1
# --- END MODIFICATION ---
plt.xlabel('Epoch')
plt.ylabel('Logloss')
plt.grid(True)
plt.legend(loc='lower right')
plt.title('Logloss over Epochs (was Accuracy plot)')

# Loss plot
plt.subplot(1, 2, 2)
# --- BEGIN MODIFICATION ---
plt.plot(history.evals_result()['validation_1']['logloss'], 'o-', label='Validation Loss')
plt.plot(history.evals_result()['validation_0']['logloss'], 'o-', label='Training Loss')
# --- END MODIFICATION ---
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend(loc='upper right')
plt.title('Loss over Epochs')

plt.show()
```
